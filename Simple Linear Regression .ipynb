                                        LINEAR REGRESSION MODEL REPRESENTATION

```python
#IMPORTING LIBRARIES
import numpy as np

import matplotlib.pyplot as plt
```

```python
# "x" VALUE IS UNIVARIATE SIZE OF HOUSE IN 1000 sq FEET

# "Y" VALUE IS TARGET PRICE OF HOUSE IN 10000 $

x=np.array([0.5,2.,3.])

y=np.array([150.,400.,600.])

print(f"The value of training set of house size are{x}")

print(f"The value of target price for  house size is {y}")
```

```python
#NUMBER OF TRAINING EXAMPLES "m"

m=len(x)

print(f"The no of training examples are {m}")
```

```python
plt.plot(x , y , "o r" )

plt.xlabel("Size of House in SQ FEET")

plt.ylabel("Price in 10000$")

plt.title("Housing Price")
```

```python
#MODEL FUNC VALUE
w=180

b=60

print(f'The value of w :{w}' )

print(f'The value of b :{b}' )
```

```python
# MODEL OUTPUT FUNCTION
def output_model(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range (m):
        f_wb[i]=w*x[i]+b
    return f_wb    
```

```python
temp_f_wb=output_model(x,w,b)
plt.plot(x,temp_f_wb, label="Our preduction")
plt.plot(x,y,"or",label="Actual value")
plt.xlabel("Size (1000 sqft)")
plt.ylabel('Price (in 1000s of dollars)')
plt.title("Housing Prices")
plt.legend()
```

```python
#New prediction

w=180

b=60

new_x=1.75

new_size_price=w*new_x+b

print(f"The price of of new house with size {new_x} in 1000's feet is {new_size_price} in 10000$")
```
